# Logistic regression

```python
# Importing various libraries
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('../apple_quality.csv')
df
# The dataset cannot be shown in its entirety because it is too large to be displayed in a single page.
# In the polynomial regression model we only have ≈ 50 rows, here we have 4001 rows.
```

```python
df.tail(1)
# We have an entire row with 'NaN's
```

```python
# Removing this row using 'drop()' function
df.drop(4000, axis=0, inplace=True)
```

```python
# Dividing the dataset into dependent and independant variables.
#   Dependant: Quality
#   Independant: All of the others except "A_id"

X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
# Since we have one column with a string we encode this column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y)
y_enc = le.transform(y)
```

```python
# Splitting the dataset into training and testing. Providing 80% for trainging and using a random state of 42
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_enc, train_size=0.2, random_state=42)
```

```python
# Instantiating the model of type logistic regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42)
```

```python
# Training the model on the test data i previously split
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cnf_matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Confusion Matrix:\n" ,cnf_matrix)

print("\n\nClassification report:\n", report)
```

# Confusion metrics

# Classification report
**Precision**: Calculates the accuracy of the positive results
* (True positives)/(True Positives+False positives)

**Recall**: Identifies the models ability to correctly predict the result
* (True positives)/(True Positives+False negatives)

**F1-score**: Creates a mean of precision and recall. When we want to consider both
* 2 * (Precision*Recall)/(Precision+Recall)
     
They relate to the CNF matrix in the way that they use the different columns in the matrix to calculate the different values.
